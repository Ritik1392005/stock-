Question 1: Extracting Tesla Stock Data (from CSV)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Tesla stock data
tesla_df = pd.read_csv("tesla.csv")

# Clean column names
tesla_df.columns = tesla_df.columns.str.strip()

# Convert 'Date' to datetime
tesla_df['Date'] = pd.to_datetime(tesla_df['Date'])
price_cols = ['Close/Last', 'Open', 'High', 'Low']
for col in price_cols:
    tesla_df[col] = tesla_df[col].replace({'\$': '', ',': ''}, regex=True).astype(float)
tesla_df['Volume'] = tesla_df['Volume'].replace({',': ''}, regex=True).astype(int)
tesla_df = tesla_df.sort_values("Date")
tesla_df.head()
```

Question 5 create a tesla stock and revenue dashboard

```python
import pandas as pd
import matplotlib.pyplot as plt
tesla_data = pd.read_csv("tesla.csv")
# Convert 'Date' column to datetime format and sort in ascending order
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'])
tesla_data = tesla_data.sort_values('Date')
tesla_data['Close/Last'] = tesla_data['Close/Last'].replace('[\$,]', '', regex=True).astype(float)
plt.figure(figsize=(12, 6))
plt.plot(tesla_data['Date'], tesla_data['Close/Last'], label='Tesla Stock Price', color='blue', linewidth=2)
plt.title("Tesla Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Question 6 - GameStop Stock and Revenue Dashboard-

```python
import pandas as pd
import matplotlib.pyplot as plt
gamestop_data = pd.read_csv("Game.csv")
gamestop_data['Date'] = pd.to_datetime(gamestop_data['Date'])
gamestop_data = gamestop_data.sort_values('Date')
gamestop_data['Close/Last'] = gamestop_data['Close/Last'].replace('[\$,]', '', regex=True).astype(float)
plt.figure(figsize=(12, 6))
plt.plot(gamestop_data['Date'], gamestop_data['Close/Last'], label='GameStop Stock Price', color='purple', linewidth=2)
plt.title("GameStop Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
